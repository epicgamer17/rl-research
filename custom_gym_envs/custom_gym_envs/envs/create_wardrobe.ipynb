```python
# [item_id, r_primary, g_primary, b_primary, r_secondary, g_secondary, b_secondary, r_tertiary, g_tertiary, b_tertiary, patterend (0 = solid, 1 = striped, 2 = repeated_images (floral tee), 3 = other), graphic_location (0 = none, 1 = front, 2 = back, 3 = both), collared_shirt (0 = not_collared, 1 = collared), length (0 = short short, 1 = short, 2 = long), body_part (0 = feet, legs, torso, accessory), fit (0 = normal, 1 = tight, 2 = oversized), material = (0 = cotton, 1 = polyester, 2 = leather, 3 = knitted, 4 = denim, 5 = other)]

import pandas as pd
import os


def add_item():
    if os.path.exists("./wardrobe.csv"):
        current_wardrobe = pd.read_csv("./wardrobe.csv")
    else:
        current_wardrobe = pd.DataFrame(
            columns=[
                "item_id",
                "r_primary",
                "g_primary",
                "b_primary",
                "r_secondary",
                "g_secondary",
                "b_secondary",
                "r_tertiary",
                "g_tertiary",
                "b_tertiary",
                "patterned",
                "graphic_location",
                "collared_shirt",
                "length",
                "body_part",
                "fit",
                "material",
            ]
        )

    item = {}
    print("current_wardrobe: ", current_wardrobe)
    item_id = current_wardrobe["item_id"].max() + 1 if not current_wardrobe.empty else 0
    print(item_id)
    r_primary = int(input("Enter the primary color's red value: "))
    g_primary = int(input("Enter the primary color's green value: "))
    b_primary = int(input("Enter the primary color's blue value: "))
    has_secondary = input("Does the item have a secondary color? (y/n): ")
    if has_secondary == "y":
        r_secondary = int(input("Enter the secondary color's red value: "))
        g_secondary = int(input("Enter the secondary color's green value: "))
        b_secondary = int(input("Enter the secondary color's blue value: "))
    else:
        r_secondary = -1
        g_secondary = -1
        b_secondary = -1
    has_tertiary = input("Does the item have a tertiary color? (y/n): ")
    if has_tertiary == "y":
        r_tertiary = int(input("Enter the tertiary color's red value: "))
        g_tertiary = int(input("Enter the tertiary color's green value: "))
        b_tertiary = int(input("Enter the tertiary color's blue value: "))
    else:
        r_tertiary = -1
        g_tertiary = -1
        b_tertiary = -1
    patterned = int(
        input(
            "Enter the pattern type (0 = solid, 1 = striped, 2 = repeated_image, 3 = other): "
        )
    )
    graphic_location = int(
        input("Enter the graphic location (0 = none, 1 = front, 2 = back, 3 = both): ")
    )
    collared_shirt = input("Is the item a collared shirt? (y/n): ")
    if collared_shirt == "y":
        collared_shirt = 1
    else:
        collared_shirt = 0
    length = int(input("Enter the length (0 = short short, 1 = short, 2 = long): "))
    body_part = int(
        input("Enter the body part (0 = feet, 1 = legs, 2 = torso, 3 = accessory): ")
    )
    fit = int(input("Enter the fit (0 = normal, 1 = tight, 2 = oversized): "))
    material = int(
        input(
            "Enter the material (0 = cotton, 1 = polyester, 2 = leather, 3 = knitted, 4 = denim, 5 = other): "
        )
    )
    item["item_id"] = item_id
    item["r_primary"] = r_primary
    item["g_primary"] = g_primary
    item["b_primary"] = b_primary
    item["r_secondary"] = r_secondary
    item["g_secondary"] = g_secondary
    item["b_secondary"] = b_secondary
    item["r_tertiary"] = r_tertiary
    item["g_tertiary"] = g_tertiary
    item["b_tertiary"] = b_tertiary
    item["patterned"] = patterned
    item["graphic_location"] = graphic_location
    item["collared_shirt"] = collared_shirt
    item["length"] = length
    item["body_part"] = body_part
    item["fit"] = fit
    item["material"] = material

    current_wardrobe.loc[len(current_wardrobe)] = item
    current_wardrobe.to_csv("wardrobe.csv", index=False)
    print("Item added successfully!")
    print("Item: ", item)
    print(current_wardrobe)
    return
```

```python
add_item()
```
